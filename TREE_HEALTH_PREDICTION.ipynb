```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import folium
url_data ='https://docs.google.com/spreadsheets/d/e/2PACX-1vSj0xe5RxZTepYQPUUKthCZdA970LFP6jo4kzOZis1SXVT_psNz0uW2xwOnuMQFGQ/pub?gid=461116058&single=true&output=csv'
data = pd.read_csv(url_data)
data.head(3)
```

```python
data.info()
```

```python
data = data.drop(['S/N', 'AUTHORITIES', 'COMMON NAME', 'FAMILIES'], axis=1)

CS =(data['CONSERVATION STATUS'].astype(str).str.strip().str.upper())
Cs= data['CONSERVATION STATUS'].value_counts()

SPP_counts= (data['SPECIES'].astype(str).str.strip().str.upper())
SPP_counts = data['SPECIES'].value_counts()

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['CONSERVATION STATUS_encoded'] = le.fit_transform(data['CONSERVATION STATUS'])
data['SPP_counts'] = le.fit_transform(data['SPECIES'])
data.head(5)


data = data.replace({'YES': 1})
data = data.replace({'NO': 0})
data = data.replace({'NO ': 0})

data.head(5)
data =data.drop(['CONSERVATION STATUS', 'SPECIES'], axis=1)

data = data[[col for col in data.columns if col != 'Healthy'] + ['Healthy']]
data.head(3)
```

TRAINING DATA

```python
# Convert 'YES' and 'NO' to 1 and 0 using the existing mapping
mapping = {'YES': 1, 'NO': 0}
for col in ['Presence of epiphytes', 'Debarking', 'Presence of Knots/bulges', 'Presence of Climbers', 'Decolourized leaves', 'Presence of epicormic branches', 'Presence of holes', 'Insect Attack']:
    data[col] = data[col].replace(mapping)

x = data.drop(['Healthy'], axis=1)
y = data['Healthy']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn import tree
from sklearn.metrics import mean_squared_error, r2_score,accuracy_score,classification_report,confusion_matrix

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size =0.2, random_state= 40 )
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)

# Make predictions on the test set
y_pred = model.predict(x_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
print(confusion_matrix(y_test, y_pred))
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)


mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
print(f"Accuracy: {accuracy}")
```

```python
model = SVC(kernel='linear',)
model.fit(x_train, y_train)

# Make predictions on the test set
y_pred = model.predict(x_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
print(f"Accuracy: {accuracy}")
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
